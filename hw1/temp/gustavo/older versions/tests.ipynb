```python
def print_tree(tree_string):
    tree_string = tree_string.strip()
    tree = nltk.Tree.fromstring(tree_string)
    tree.pretty_print()

def draw_tree(tree_string):
    tree_string = tree_string.strip()
    tree = nltk.Tree.fromstring(tree_string)
    tree.draw()

def get_rules(tree):
    # Get subtrees of the tree
    subtrees = tree.subtrees()
    rules = []

    # For each subtree get the rule that generates it's immediate child node
    for subtree in subtrees:
        children = [ch for ch in subtree]

        # Skip preterminals
        if len(children) > 1:
            # get labels of child nodes
            child_labels = [ch.label() for ch in children]

            # create rule string
            curr_rule = ' '.join([subtree.label()] + child_labels)
            rules.append(curr_rule)

    return rules
```

```python
def get_tags(word, preterminals):
    tags = []
    for pt in preterminals:
        leaf = [child for child in pt][0]

        if word == leaf and not pt.label() in tags:
            tags.append(pt.label())

    return tags
```

```python
# Lines in the vocab file
vocab_lines =[]
unassigned_words = []
```

```python
# Get tags associated with each word in the corpus
for word in allowed_words:
    curr_tags = get_tags(word, preterminals)

    if not curr_tags:
        unassigned_words.append(word)
        continue

    for tag in curr_tags:
        vocab_lines.append('{0}    {1}    {2}'.format(1, tag, word))
```
